```python
import pandas as pd
import numpy as np

data = {
    'Firstname': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Class': [10, 11, 10, 12, 11, 10, 11, 10, 12, 11, 10, 11, 10, 12, 11, 10, 12],
    'Subject': ['Math', 'Science', 'English', 'Math', 'Science', 'Physics', 'Science', 'English', 'Math', 'Science','Science', 'Science', 'English', 'Math', 'Science', 'History', 'Physics'],
    'Score': [85, 92, 78, 88, 95, 80, 90, 82, 89, 93, 75, 87, 91, 94, 90, 79, 86]
}

df = pd.DataFrame(data)
df
```

```python
# Do group by
df_firstname = df.groupby('Firstname')

for key, dataFrame in df_firstname:
    print(key)
    print(dataFrame)
    print()
```

```python
df_firstname.mean(numeric_only=True)
```

```python
df_firstname.max(numeric_only=True)
```

```python
# prompt: Would you like to get back the code before refrshing the page

import pandas as pd
import numpy as np

data = {
    'Firstname': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Class': [10, 11, 10, 12, 11, 10, 11, 10, 12, 11, 10, 11, 10, 12, 11, 10, 12],
    'Subject': ['Math', 'Science', 'English', 'Math', 'Science', 'Physics', 'Science', 'English', 'Math', 'Science','Science', 'Science', 'English', 'Math', 'Science', 'History', 'Physics'],
    'Score': [85, 92, 78, 88, 95, 80, 90, 82, 89, 93, 75, 87, 91, 94, 90, 79, 86]
}

df = pd.DataFrame(data)
df

# Do group by
df_firstname = df.groupby('Firstname')

for key, dataFrame in df_firstname:
    print(key)
    print(dataFrame)
    print()

df_firstname.mean(numeric_only=True)
#df_firstname.max(numeric_only=True)
```

```python
df_firstname_subject = df.groupby('Firstname')['Subject'].value_counts()
print(df_firstname_subject)

total_subject_by_firstname = df.groupby('Firstname')['Subject'].count()
print(total_subject_by_firstname)

# df.groupby(['Firstname', 'Class', 'Subject']).mean()
```

```python
# Sample dataset
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'gender': ['female', 'male', 'male', 'female', 'female', 'male']
}

df = pd.DataFrame(data)

# count value of gender
count_values_gender = df['gender'].value_counts()
print("Count values of gender",count_values_gender)

# number of name by gender
names_by_gender = df.groupby(['gender'])['name'].count()
print("Names by gender",names_by_gender)

# number of gender by name
genders_by_name = df.groupby(['name'])['gender'].count()
print("Genders by names",genders_by_name)
```
