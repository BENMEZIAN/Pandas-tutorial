# Example

```python
Student = {
    'Name' : ['Mohamed','Amine','Aymen','Kader'],
    'Std' : [9,10,9,10],
    'Age' : [20,21,20,21],
    'Fav subject' : ['DS','DSA','WD','CS']
};
```

```python
df_student = pd.DataFrame(Student)
df_student
```

```python
df_student['Name'] # select one column
# type(df_student['Name']) # pandas.core.series.Series
```

```python
df_student[['Std','Name','Fav subject']] # select multiple columns
# type(df_student[['Std','Name','Fav subject']]) # pandas.core.frame.DataFrame
```

### Select row(s) and column(s)

```python
# How to select row(s) and column(s)
## iloc : integer locate
# Syntax : df.iloc[row(int),column(int)]
print("iloc[0] : ",df_student.iloc[0]) # first row
print("--------------------------------------------")
print("iloc[1] : ",df_student.iloc[1]) # second row
print("--------------------------------------------")
print(df_student.iloc[0:2]) # first two rows
print("--------------------------------------------")
print(df_student.iloc[0:2,0:2]) # first two rows and first two columns
print("--------------------------------------------")
print("iloc[1,3] : ",df_student.iloc[1,3]) # DSA
print("--------------------------------------------")
print("iloc[2,2] : ",df_student.iloc[2,2]) # 20
print("--------------------------------------------")
print("iloc[3,3] : ",df_student.iloc[3,3]) # CS
print("--------------------------------------------")
```

```python
## loc : locate
# Syntax : df.loc[row(index),column(name)]
print("first row : ",df_student.loc[0]) # first row
print("--------------------------------------------")
print("first two rows : ",df_student.loc[0:1,:]) # first two rows
print("--------------------------------------------")
print("first two rows : ",df_student.loc[0:1,['Name','Fav subject']]) # first two rows with subset columns
print("--------------------------------------------")
print("All rows : ",df_student.loc[:,:]) # All rows
```
