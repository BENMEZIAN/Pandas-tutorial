```python
#Checking for missing values
print(df.isnull())

# Checking for non-missing values
print(df.notnull())
```

```python
# Checking missing values methods
print(df.isnull().sum()) # missing values at each column
print(df.isnull().sum().sum()) # total missing values
```

```python
# How to handle them (dropping or replacing)
# fillna() method, fill with median, mean, mode

for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'float64' or df[col].dtype == 'int64':
            df[col].fillna(df[col].mean(), inplace=True)


for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)

df

# Checking for mimssing values
print(df.isnull().sum())
print(df.isnull().sum().sum())
```
