```python
import pandas as pd
```

```python
# Create student dictionnary
Student = {
    'Name' : ['Mohamed','Amine','Aymen','Kader'],
    'Std' : [9,10,9,10],
    'Age' : [20,21,20,21],
    'Fav subject' : ['DS','DSA','WD','CS']
};

df_student = pd.DataFrame(Student)
df_student
```

```python
# Do some examples on this dataset
# rename():
df_student.rename(columns={'Name': 'First Name',
                           'Fav subject': 'Favorite subject'},
                  inplace=True)
df_student
```

```python
# Add columns
df_student['Gender'] = ['M','M','M','M']
df_student['Last Name'] = ['Japoni','Bennani','Benali','Kada']
df_student['Full Name'] = df_student['First Name'] + ' ' + df_student['Last Name']
df_student
```

```python
# Drop columns
df_student.drop(columns=['First Name', 'Last Name'], inplace=True)
df_student
```

```python
# Reoder columns
df_student = df_student[['Full Name', 'Gender', 'Std', 'Age', 'Favorite subject']]
df_student
```

```python
# Pandas Columns: Renaming, Adding, and Dropping Columns
#
# 1. **Renaming Columns**:
#    - You can rename columns using the `rename()` function with the `columns` parameter.
#    - `rename()` returns a new DataFrame with updated column names.
#    - Use `inplace=True` to modify the DataFrame directly.
#
# Example:
# df.rename(columns={'old_name': 'new_name'}, inplace=True)
#
# 2. **Adding Columns**:
#    - To add a new column, simply assign a value or a list to a new column name.
#    - You can add a column with a constant value, or from an existing column (e.g., performing calculations).
#
# Example:
# df['new_column'] = [value1, value2, value3, ...]  # Adding a constant column
# df['age_plus_5'] = df['Age'] + 5  # Adding a derived column
#
# 3. **Dropping Columns**:
#    - To drop a column, use the `drop()` method with `axis=1` (to specify columns).
#    - By default, `drop()` returns a new DataFrame without the dropped column unless `inplace=True` is used.
#    - To remove multiple columns, pass a list of column names.
#
# Example:
# df.drop(columns=['column_name'], inplace=True)  # Drop a single column
# df.drop(columns=['column1', 'column2'], inplace=True)  # Drop multiple columns
#
# 4. **Summary of Functions**:
#    - `rename(columns={})`: Rename one or more columns.
#    - `df['new_column'] = value`: Add a new column.
#    - `drop(columns=[...])`: Drop one or more columns.
#
# These operations help in managing DataFrame structures by modifying columns to suit your analysis needs.
```
