```python
import pandas as pd
```

```python
# Sample student data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [18, 19, 17, 20, 18],
        'Grade': ['A', 'B', 'C', 'A', 'B']}

# Create a DataFrame
student_df = pd.DataFrame(data)

# Display the DataFrame
student_df
```

```python
# Sort dataset by one column
sorted_df_ascending = student_df.sort_values(by='Age') # By default in ascending order
sorted_df_descending = student_df.sort_values(by='Age', ascending=False) # By default in ascending order
#student_df.sort_values(by='Age', ascending=False,inplace=True) # change the original dataset with 'inplace' parameter

print("Ascending sorted :\n",sorted_df_ascending)
print("Descending sorted :\n",sorted_df_descending)
```

```python
# Sort dataset by multiple columns
sorted_df_multiple = student_df.sort_values(by=['Age', 'Grade'], ascending=[True, True])
print("Multiple columns sorted :\n",sorted_df_multiple)
```

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}

df = pd.DataFrame(data)

# Sorting by 'Score' in ascending order
sorted_df = df.sort_values(by=['Age', 'Score'])
sorted_df
```

```python
# prompt: give me another example to sort an important dataset amout with sort_values

import pandas as pd

# Sample data with more realistic values and potential for ties
data = {'StudentID': [101, 102, 103, 104, 105, 106, 107],
        'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Bob', 'Alice'],
        'Major': ['Computer Science', 'Physics', 'Mathematics', 'Computer Science', 'Biology', 'Physics', 'Engineering'],
        'GPA': [3.8, 3.5, 3.9, 3.8, 3.7, 3.5, 3.2],
        'EnrollmentDate': pd.to_datetime(['2022-09-01', '2022-09-05', '2022-08-28', '2022-09-01', '2022-08-25', '2022-09-10','2023-01-15'])}


student_df = pd.DataFrame(data)

# Sort by multiple columns: GPA (descending), then EnrollmentDate (ascending)
sorted_students = student_df.sort_values(by=['GPA', 'EnrollmentDate'], ascending=[False, True])

print(sorted_students)
```

```python
#Example of sorting with NaNs
import numpy as np
data_with_nan = {'StudentID': [101, 102, 103, 104, 105],
        'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'GPA': [3.8, 3.5, np.nan, 3.8, 3.7]}

df_with_nan = pd.DataFrame(data_with_nan)
sorted_df_nan = df_with_nan.sort_values(by='GPA', ascending=False, na_position='first') # NaN values will be at the beginning
print("\nDataFrame with NaN values sorted:\n",sorted_df_nan)
```
