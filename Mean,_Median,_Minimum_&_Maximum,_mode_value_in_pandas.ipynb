```python
import pandas as pd

# Change data dictionnary into only numerical columns
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve','Alice'],
    'Age': [25, 30, 22, 28, 35,30],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Sydney','London'],
    'Salary': [50000, 60000, 55000, 70000, 80000, 90000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR','IT']
}

student_df = pd.DataFrame(data)
student_df
```

```python
# Calculate mean, median, max, min for numeric columns only
numeric_cols = student_df.select_dtypes(include='number')  # Select only numeric columns

sum_values = numeric_cols.sum()
mean_values = numeric_cols.mean()
median_values = numeric_cols.median()
max_values = numeric_cols.max()
min_values = numeric_cols.min()
mode_values = numeric_cols.mode(numeric_only=True).iloc[0]

# Display the results
print("Sum values (numeric):\n", sum_values)
print("----------------------------------------")
print("Mean values (numeric):\n", mean_values)
print("----------------------------------------")
print("Median values (numeric):\n", median_values)
print("----------------------------------------")
print("Max values (numeric):\n", max_values)
print("----------------------------------------")
print("Min values (numeric):\n", min_values)
print("----------------------------------------")
print("Mode values (numeric):\n", mode_values)
print("----------------------------------------")
```

```python
# For string columns, we can calculate the mode (most frequent value)
string_cols = student_df.select_dtypes(include='object')  # Select only string columns
print(string_cols)

# Loop through each string column and calculate mode
for str_col in string_cols:
    mode_value = student_df[str_col].mode()  # Mode for the current column
    print(f"Mode value for column '{str_col}': {mode_value.iloc[0] if not mode_value.empty else 'No Mode'}")
```

```python
# Sum applied on rows
# Select only numeric columns (excluding string columns)
numeric_cols = student_df.select_dtypes(include=['number'])

# Sum the rows across numeric columns only
Age_salary_sum = numeric_cols.sum(axis=1)

# add this sum to dataframe
student_df['Age_salary_sum'] = Age_salary_sum
student_df
```
