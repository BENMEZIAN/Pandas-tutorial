```python
import pandas as pd
```

```python
# Create student dictionnary
Student = {
    'Name' : ['Mohamed','Amine','Aymen','Kader'],
    'Std' : [9,10,9,10],
    'Age' : [20,21,20,21],
    'Fav subject' : ['DS','DSA','WD','CS']
};
```

```python
df_student = pd.DataFrame(Student)
df_student
```

```python
filtered = df_student['Std'] == 10
```

```python
# Filter rows simply
df_student[filtered]
```

```python
df_student.loc[filtered,['Name','Fav subject','Age']]
```

```python
f = df_student['Name'].str.contains('Amine')
df_filtered = df_student[f]
df_filtered
```

```python
# Read dataset
df = pd.read_csv("heroes_information.csv")
df
```

```python
# Dataset subset with condition which is publisher == 'Marvel Cosmics' (== equal) operator
f = df['Publisher'] == 'Marvel Comics'
df_heros = df[f]
df_heros
```

```python
# Condition with & (and) operator
f2 = (df_heros['Gender'] == 'Female') & (df_heros['Publisher'] == 'Marvel Comics')
df_heros[f2]
```

```python
# Condition with | (or) operator
f3 = (df_heros['Race'] == 'Human') | (df_heros['Race'] == 'Mutant')
df_heros[f3]
```

```python
# prompt: isin function in pandas which verify the existing of any color on tha dataset of the colors existing in an array of colors
colors = ['red', 'green', 'blue','black','white','purple']
f4 = df_heros['Eye color'].isin(colors)
df_heros[f4]
```

```python
# Condition with != (not equal) operator
f5 = df_heros['Skin color'] != '-'
df_heros[f5]
```
