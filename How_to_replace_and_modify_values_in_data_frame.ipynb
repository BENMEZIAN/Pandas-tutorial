```python
# prompt: Use apply() with lambda function, map() function, applymap()

import pandas as pd
# Create student dictionnary
Student = {
    'Name' : ['Mohamed','Amine','Aymen','Kader'],
    'Std' : [9,10,9,10],
    'Age' : [20,21,20,21],
    'Fav subject' : ['DS','DSA','WD','CS'],
    'Grade': ['A', 'B', 'A', 'C']
};

df_student = pd.DataFrame(Student)
print(df_student)

# Apply() with lambda function
df_student['Age'] = df_student['Age'].apply(lambda x: x + 1)

# map() function
df_student['Std'] = df_student['Std'].map({9: 10, 10: 11})
df_student['Fav subject'] = df_student['Fav subject'].map({'DS': 'Data Science', 'DSA': 'Data Structures and Algorithms', 'WD': 'Web Development', 'CS': 'Computer Science'})
df_student['Grade'] = df_student['Grade'].map({'A': 'Excellent', 'B': 'Good', 'C': 'Average'})

# applymap()
# Example: Convert all string values to uppercase (if applicable)
# Note that this will affect the entire DataFrame
# df_student = df_student.applymap(lambda x: x.upper() if isinstance(x, str) else x)
# Give note (between 0 and 20) according to the Grade column and add it to the dataset
def give_note(grade):
    if grade == 'Excellent':
        return 20
    elif grade == 'Good':
        return 15
    elif grade == 'Average':
        return 10

df_student.loc[1,'Fav subject'] = 'AI'
df_student.loc[3] = ['Ali',10,20,'ML','Average']
df_student.loc[2,['Std','Fav subject']] = [11,'Web Dev']
df_student.replace({9:10,10:11,20:21},inplace=True)
df_student['Age'].replace(21,22)
df_student['Std'].replace(10,11)
df_student.rename(columns={'Name':'First Name'},inplace=True)
df_student['Last Name'] = ['Benali','benz','kada','barakoda']
df_student['Full Name'] = df_student['First Name'] + ' ' + df_student['Last Name']
df_student.drop(columns=['Last Name'],inplace=True)
df_student['Full Name'] = df_student['Full Name'].str.upper()
df_student['Note'] = df_student['Grade'].apply(give_note) # give note is a function
df_student = df_student[['Full Name','Std','Age','Fav subject','Note','Grade']]
df_student
```

```python
# Pandas: Replacing and Modifying Values in a DataFrame
#
# 1. **Replacing Values**:
#    - Use `replace()` to replace specific values in the DataFrame.
#    - You can replace single or multiple values in a column or the entire DataFrame.
#    - You can replace values conditionally or globally.
#
# Example (Replace a value in the entire DataFrame):
# df.replace(1, 100, inplace=True)  # Replace all occurrences of '1' with '100'
#
# Example (Replace multiple values):
# df.replace({1: 100, 'A': 'Z'}, inplace=True)  # Replace 1 with 100, 'A' with 'Z'
#
# Example (Replace values in a specific column):
# df['column_name'].replace(1, 100, inplace=True)  # Replace '1' with '100' in a specific column
#
# 2. **Modifying Values Based on Condition**:
#    - Use boolean indexing or `.loc[]` to modify values based on a condition.
#    - You can modify values in one or more columns where the condition is met.
#
# Example (Modify based on condition):
# df.loc[df['Age'] > 30, 'Age'] = 40  # Set 'Age' to 40 where 'Age' is greater than 30
#
# 3. **Modifying Values Using `apply()`**:
#    - You can use the `apply()` function to apply a function across a column or row.
#    - This is useful for more complex modifications.
#
# Example (Using `apply()` to modify values):
# df['Age'] = df['Age'].apply(lambda x: x + 5)  # Add 5 to every value in the 'Age' column
#
# 4. **Modify Using `.map()` or `.replace()` (for mapping values)**:
#    - Use `.map()` to replace values in a column based on a dictionary or a function.
#    - `.map()` is often used to modify or replace categorical variables.
#
# Example (Modify categorical values using `.map()`):
# df['Std'] = df['Std'].map({9: 'Grade 9', 10: 'Grade 10'})  # Map '9' to 'Grade 9' and '10' to 'Grade 10'
#
# 5. **Modify with `applymap()` (for entire DataFrame)**:
#    - If you need to modify every element of the entire DataFrame, use `applymap()`.
#
# Example (Using `applymap()` to modify every element):
# df = df.applymap(lambda x: x * 2)  # Multiply every element in the DataFrame by 2
#
# 6. **Summary of Functions**:
#    - `replace()`: Replace specific values in the DataFrame.
#    - `loc[]`: Modify values based on conditions.
#    - `apply()`: Apply a function along a column or row to modify values.
#    - `map()`: Map values in a column using a dictionary or function.
#    - `applymap()`: Modify all values in the DataFrame.
#
# These operations are helpful for data cleaning, transformation, and feature engineering.
```
